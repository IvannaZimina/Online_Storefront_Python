## Online Storefront — Run & Test Commands (for reviewers)

#### 1) Requirements (must be installed on the machine)
- **Python 3.10+**
- **Git**
- **(recommended) VS Code** with Python extension
- Internet access to install dependencies
> If any command below fails, first ensure the virtual environment is **activated** and the project is **installed** as editable (Sections 3–4).

#### 2) Clone the repository
```bash
# choose any directory on your machine and run:
git clone <REPOSITORY_URL>
cd <REPOSITORY_DIR>   # enter the cloned project folder
```

#### 3) Create and activate the virtual environment
All subsequent commands are executed from the project root (where `src/`, `tests/`, and `pyproject.toml` are located).
**Create venv (once):**
```bash
python -m venv venv
```
**Activate venv (every new terminal session):**
• **Git Bash / WSL / bash:**  `source venv/Scripts/activate`
• **PowerShell:**             `venv\Scripts\Activate`
You should see `(venv)` at the start of your terminal prompt.

#### 4) Install the project (required for imports to work)
The project uses a `src/` layout. Install it in editable mode so Python can import all domain packages:
```bash
pip install -e .
```
If this command fails with “no pyproject.toml or setup.py”:
You are not in the project root.  
`cd` into the directory that contains `pyproject.toml` and retry.
**Expected success message (example):**
Successfully installed online-storefront-<version>

### 5) Install test tools
Install pytest:
```bash
pip install pytest
```
**optional coverage support:**
```bash
pip install pytest-cov
```
If python `-m pytest` reports `“No module named pytest”`:
You either forgot to install pytest or the venv is not active (activate it and run the install again).

#### 6) Run the domain demo (CLI)
A small CLI demo exists for the product domain:
```bash
python -m product.cli
```
(If demos for other domains are added, they run similarly:)
```bash
python -m cart.cli
python -m payment.cli
python -m shipping.cli
```
If running python `-m product.cli` fails with `ModuleNotFoundError: product`, confirm Sections 3–4 were completed (venv active, project installed).

#### 7) Run all tests
```bash
python -m pytest
```
With coverage:
```bash
python -m pytest --cov
```
Do not run test files as plain scripts (e.g., `python tests/test_product.py`) — with the src/ layout this often breaks imports.
Always use `python -m pytest …` or the alternative below.
To run tests with the --cov option, use the following commands:
For check
```bash
pip list
```
If there is no pytest-cov, install it:
```bash
pip install pytest-cov
```
Then use the command:
```bash
python -m pytest --cov
```
This will run all tests and generate a coverage report in the terminal.

#### 8) Run a single test file (or test module)
```bash
python -m pytest tests/test_product.py
python -m pytest tests/test_product_service.py
```
**examples for other domains:**
```bash
python -m pytest tests/test_cart.py
python -m pytest tests/test_payment.py
```
Alternative without pytest options (works if the project is installed):
```bash
python -m tests.test_product
python -m tests.test_product_service
```
#### 9) Useful maintenance commands
Reinstall the project after structural changes (e.g., new packages under src/)
```bash
pip install -e .
```
Show what’s installed in the virtual environment
```bash
pip list
```
Deactivate the virtual environment
```bash
deactivate
```

#### 10) Troubleshooting (quick answers)
**A) ModuleNotFoundError: No module named 'product'** (or cart/payment/shipping)
•	Ensure the virtual environment is activated (Section 3).
•	Ensure the project is installed in editable mode (Section 4): `pip install -e .`
•	Run tests/demos via module execution: 
```bash
python -m pytest
python -m product.cli
```
**B) No module named pytest**
•	Install pytest inside the active venv: pip install pytest
**C) VS Code import squiggles (editor shows unresolved imports)**
•	Optionally add .env in the project root: `PYTHONPATH=src`
•	And in `.vscode/settings.json: { "python.envFile": "${workspaceFolder}/.env" }`
(This improves editor linting; runtime still requires Sections 3–4.)
